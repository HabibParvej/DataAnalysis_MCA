```python
data = pd.read_csv('/content/train.csv')
data.head()
```

# Scatterplot
# A scatter plot of y vs. x with varying marker size and/or color.
# It plots relationship of two features.

```python
plt.scatter(x=data['km'], y=data['current price'],color='red')
plt.show()
```

```python
plt.scatter(x=data['km'], y=data['current price'])
plt.title('Scatter Plot')
plt.xlabel('km')
plt.ylabel('current price')
plt.show()
```

```python
dept = ('BCA', 'BBA', 'MCA')
noStudent = (120, 82, 50)
plt.bar(dept, noStudent)
```

```python
# Plotting for No of Car For Different RAting
car = ('suzuki', 'ferrari', 'jeep')
rating = (2, 5, 4)
plt.bar(car, rating, width=0.5,color=['black','black','black'])
```

```python
dept=['BCA', 'BBA', 'MCA']
noStudentPresent=[100,50,40]
plt.bar(dept,noStudent,width=0.3,label='No_Students')
plt.bar(dept,noStudentPresent,width=0.3,bottom=noStudent,label='No_Students_Present')
plt.title('Department wise student and present')
plt.xlabel('Department')
plt.ylabel('No of students')
plt.legend()

```

```python
x=[1,3,5,8,9,20,25]
plt.plot(x)
plt.xlim(-2,10)
plt.ylim(-2,30)
plt.xticks(np.arange(-2,10,1.5))
plt.title('Autoscaling Plot')
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 5*np.pi, 0.1)
sin1 = np.sin(x)
cos1 = np.cos(x)
tan1 = np.tan(x)
lg = np.log(x)


# Plot data on individual subplots
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(x, sin1)
plt.title('Sin Plot')
plt.subplot(2,2,2)
plt.plot(x, tan1)
plt.title('tan Plot')
plt.subplot(2,2,3)
plt.plot(x, cos1)
plt.title('cos Plot')
plt.subplot(2,2,4)
plt.plot(x, lg)
plt.title('Logarithm Plot')
 # Adjust the layout
plt.tight_layout()
plt.suptitle('My Graphs')
# Display the Multiplots
plt.savefig('graph.pdf')
plt.show()
```
