```python
# Seaborn
# behind the scene seaborn use matplotlib to draw its plots.
import seaborn as sns
sns.set_theme()
```

```python
dataset.head()
```

```python
dataset.size
dataset.info
```

```python
sns.scatterplot(data=dataset,x=dataset['horsepower'],y=dataset['mpg'])
```

```python
sns.scatterplot(data=dataset,x=dataset['horsepower'],y=dataset['mpg'],hue='cylinders',size='model_year')
```

```python
data = sns.load_dataset('mpg')
sns.kdeplot(data=data, x='mpg')
sns.rugplot(data=data, x='mpg')
```

```python
sns.regplot(data=data, x='horsepower',y='mpg')
```

```python
sns.countplot(data=data, x='cylinders')
```

```python
sns.countplot(data=data, x='cylinders',hue='origin')
```

```python
sns.boxplot(data=data,x='mpg')
```

```python
sns.boxplot(data=data,x='mpg',y='origin')
```

```python
sns.boxplot(data=data,x='mpg',y='origin',hue='cylinders')
```

```python
sns.boxplot(data=data,x='mpg',y='origin',hue='cylinders',notch=True)
```

```python
#Violinplot
#Boxplot with distribution and density
sns.violinplot(data=data,x='mpg')
```

```python
#Violinplot
#Boxplot with distribution and density
sns.violinplot(data=data,x='mpg',y='origin')
```

```python
sns.violinplot(data=data,x='mpg',y='origin',hue='cylinders')
```

```python
#Pair Plot
sns.pairplot(data)
```

```python
d_new=data[['displacement','horsepower','acceleration']]
d_new.head()
```

```python
corr=d_new.corr()
print(corr)
```

```python
sns.heatmap(corr)
```

```python
sns.heatmap(corr,vmin=1,vmax=1,annot=True,linewidth=0.5)
```
